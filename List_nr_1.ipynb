* a) make a 2d plot containing 10 points: $(3,5), (4,6), (7,7), (8,11), (10,12), (11,14), (12, 10), (16,15), (18, 19) , (20, 21)$
* b) add to the plot a line $y=0.6x+5$
* c) for each data $(x_i,y_i)$ point compute its distance to the line $y=0.6x+5$ defined simply as $dist((x_i,y_i), line)=(0.6x_i+5 - y_i)^2$
* d) try to  find $a,b$ such that sum of distances of all points from line $y=ax+b$ is smaller

```python
def distance(x,y,a,b):
  return sum([(a*x[i] + b - y[i])**2 for i in range(0, len(x))])

x = [3,4,7,8,10,11,12,16,18,20]
y = [5,6,7,11,12,14,10,15,19,21]

plt.scatter(x,y)
plt.axline(xy1 = (0,5), slope = 0.6, color = 'red')
plt.axline(xy1 = (0,2.199), slope = 0.9, color = 'pink')
plt.show()

dist_org = [(0.6*x[i] + 5 - y[i])**2 for i in range(0, len(x))]

a = [0.6+i*0.1 for i in range(-101,101)]
b = [5+i*0.1 for i in range(-101,101)]
indexes = (0,0)
min_dist = sum(dist_org)
for i in range(0, len(a)):
  for j in range(0, len(b)):
    d = distance(x,y,a[i], b[j])
    if d < min_dist:
      min_dist = d
      indexes = i,j

print(min_dist, a[indexes[0]], b[indexes[1]])
```
